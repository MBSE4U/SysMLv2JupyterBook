# Experimental Model for a Forest Fire Observation Drone

Copyright 2023 MBSE4U


```sysml
package ForestFireObservationDrone {
    
    import Drone_SharedAssetsSuperset::**;
    
    part forestFireObservationDrone : DroneProductLine {
        
        // Power System
        :>> battery = battery::longDistanceBattery;
        
        part charger;
        part powerManagementModule {
            port ctrlPort {
                out attribute status : BatteryCapacityKind;
            }
            
            perform action checkStatus {
                out attribute status : BatteryCapacityKind;
                
                first start;
                then action getBatteryStatus;
                then action analyseStatus;
                then done;
            } 
        }

        // Flight Control System
        part droneControlUnit {
            port powerCtrlPort {
                in attribute batteryStatus : BatteryCapacityKind;
            }
        }
        part gpsModule;
        part inertialMeasurementUnit;
        part barometer;
        
        // Drone Frame
        part body;
        part rotors[4];
        part arms[4];
        part landingGear;
        
        // Communication System
        part radioTransmitter;
        part radioReceiver;
        part antenna;
        
        // Sensors
        part camera;
        part obstacleAvoidanceSensor;
        
        // Engine
        part engine1 : DroneEngine {
            :>> propKind = PropKind::clockwise;
        }
        part engine2 : DroneEngine {
            :>> propKind = PropKind::counterclockwise;
        }
        part engine3 : DroneEngine {
            :>> propKind = PropKind::clockwise;
        }        
        part engine4 : DroneEngine {
            :>> propKind = PropKind::counterclockwise;
        }
        
        // Connections
        connect droneControlUnit.powerCtrlPort to powerManagementModule.ctrlPort;
        connect battery to powerManagementModule;
        connect powerManagementModule to engine1;
        connect powerManagementModule to engine2;
        connect powerManagementModule to engine3;
        connect powerManagementModule to engine4;                    
    }
}
```

# Shared Assets Superset

The shared assets superset contains all shared assets of the drone product line. The forest fire observation drone is one product of the product line. Other member products could be a camera drone, a surveillance drone, and so forth.

```sysml
library package Drone_SharedAssetsSuperset {
    
    package DroneProductLine {
        
        import DroneBattery::**;
        import DroneEngine::**;
               
        abstract part def DroneProductLine {
            
            // Power
            part battery :> droneBatteryVariation;
            
            // Engines
            variation attribute numberOfEngines : ScalarValues::Integer {
                variant attribute fourEngines : ScalarValues::Integer = 4;
                variant attribute sixEngines : ScalarValues::Integer = 6;                
            }    
            satisfy Drone_FeatureModel::drone.numberOfEngines by numberOfEngines;                
            part engine[4..numberOfEngines] : DroneEngine;       
        }        
    }
     
    package DroneEngine {
        
        package DroneEngine_SystemArchitecture {
        
            part def DroneEngine {
                enum propKind : PropKind;
            
                enum def PropKind {
                    enum clockwise;
                    enum counterclockwise;
                }            
            }
        }
        
                
        package DroneEngine_StakeholderRequirements {
            
            variation requirement droneEngineStakeholderRequirements {
                variant requirement droneEngineStandardStakeholderRequirements {
                    requirement droneEnginePower : DroneEnginePower;
                    requirement droneEngineEfficiency : DroneEngineEfficiency;
                    requirement droneEngineDurability : DroneEngineDurability;
                    requirement droneEngineReliability : DroneEngineReliability;
                    requirement droneEngineLightweight : DroneEngineLightweight;
                    requirement droneEngineControl : DroneEngineControl;
                    requirement droneEngineCompatibility : DroneEngineCompatibility;
                    requirement droneEngineSafety : DroneEngineSafety;
                    requirement droneEngineCostEffective : DroneEngineCostEffective;
                }
                variant requirement droneEngineLowNoiseStakeholderRequirements :> droneEngineStandardStakeholderRequirements {
                    requirement droneEngineLowNoise : DroneEngineLowNoise;
                }
            }
            
            requirement def <'DE-REQ-1'> DroneEnginePower {
                doc /*
                    * The engine should provide enough power to lift the drone off the ground and keep it in the air. 
                    * The power output should be appropriate for the size and weight of the drone.
                    */
            }
            requirement def <'DE-REQ-2'> DroneEngineEfficiency {
                doc /*
                    * The engine should be efficient, allowing the drone to fly for an extended period on a single charge. 
                    * The engine should be designed to minimize energy loss through friction and heat.
                    */
            }
            requirement def <'DE-REQ-3'> DroneEngineDurability {
                doc /*
                    * The engine should be able to withstand the stress of continuous use and exposure to the elements. 
                    * It should be able to operate in a wide range of temperatures and weather conditions.
                    */
            }
            requirement def <'DE-REQ-4'> DroneEngineReliability {
                doc /*
                    * The engine should be reliable, with minimal downtime or maintenance requirements. 
                    * It should be designed with high-quality components and undergo rigorous testing before deployment.
                    */
            }
            requirement def <'DE-REQ-5'> DroneEngineLowNoise {
                doc /*
                    * The engine should produce minimal noise to avoid disturbing the environment or attracting unwanted attention.
                    */
            }
            requirement def <'DE-REQ-6'> DroneEngineLightweight {
                doc /*
                    * The engine should be lightweight, allowing for maximum payload capacity and longer flight times.
                    */
            }
            requirement def <'DE-REQ-7'> DroneEngineControl {
                doc /*
                    * The engine should be easy to control, allowing for precise maneuverability and responsiveness to commands.
                    */
            }
            requirement def <'DE-REQ-8'> DroneEngineCompatibility {
                doc /*
                    * The engine should be compatible with the drone's power supply and electronic components.
                    */
            }
            requirement def <'DE-REQ-9'> DroneEngineSafety {
                doc /*
                    * The engine should be designed with safety in mind, minimizing the risk of injury or damage in case of malfunction. 
                    * It should have fail-safe mechanisms to prevent accidents and minimize damage in case of emergency.
                    */
            }
            requirement def <'DE-REQ-10'> DroneEngineCostEffective {
                doc /*
                    * The engine should be cost-effective, balancing performance and quality with affordability. 
                    * It should provide value for money, with a low cost of ownership and maintenance.            
                    */
            }            
        }
    }
    
    package DroneBattery {
        
        package DroneBattery_Variation {
        
            import DroneBattery_SystemArchitecture::*;
            
            variation part droneBatteryVariation : DroneBattery {
                variant part standardBattery : StandardDroneBattery;
                variant part longDistanceBattery : LongDistanceDroneBattery;                
            }
            satisfy Drone_FeatureModel::drone.batteryKind by droneBatteryVariation;                
        }
        
        package DroneBattery_SystemArchitecture {
        
            abstract part def DroneBattery {
                attribute maxCapacity;
                enum currentCapacityLevel : BatteryCapacityKind;
                enum def BatteryCapacityKind :> ScalarValues::Integer {
                    enum full = 100;
                    enum medium = 50;
                    enum low = 30;
                    enum critical = 15;
                }            
            }
            part def StandardDroneBattery :> DroneBattery {
                :>> maxCapacity = 23;
            }
            part def LongDistanceDroneBattery :> DroneBattery {
                :>> maxCapacity = 42;
            }
        }
        
        package DroneBattery_StakeholderRequirements {
            
            
            requirement def <'DB-REQ-1'> HighEnergyDensity {
                doc /*
                     * A drone battery should have a high energy density to provide sufficient power 
                     * to fly the drone for an extended period of time.
                    */
            }

            requirement def <'DB-REQ-2'> Lightweight {
                doc /*
                     * A drone battery should be lightweight to keep the overall weight of the drone 
                     * to a minimum, which is essential for stable flight.
                    */
            }

            requirement def <'DB-REQ-3'> HighCapacity {
                doc /*
                     * A drone battery should have a high capacity to provide the drone with enough power 
                     * to fly for a reasonable amount of time.
                    */
            }

            requirement def <'DB-REQ-4'> FastCharging {
                doc /*
                     * A drone battery should be able to charge quickly so that the drone can 
                     * get back in the air as soon as possible.
                    */
            }

            requirement def <'DB-REQ-5'> LongLife {
                doc /*
                     * A drone battery should have a long life to avoid the need for frequent replacements.
                    */
            }

            requirement def <'DB-REQ-6'> Safe {
                doc /*
                     * A drone battery should be designed to be safe, with features such as overcharge protection, 
                     * short circuit protection, and temperature control.
                    */
            }

            requirement def <'DB-REQ-7'> Compatible {
                doc /*
                     * A drone battery should be compatible with the drone's power system to ensure proper functioning.
                    */
            }
            
            requirement def <'DB-REQ-8'> Reliable {
                doc /*
                     * A drone battery should be reliable, with consistent performance and the ability to withstand 
                     * the stresses of flight and charging cycles.
                    */
            }

            requirement def <'DB-REQ-9'> Affordable {
                doc /*
                     * A drone battery should be reasonably priced, to avoid being a significant expense for drone pilots.
                    */
            }
            
            requirement def <'DB-REQ-10'> EasyToInstall {
                doc /*
                     * A drone battery should be easy to install and remove to make it easy for pilots to swap out batteries when necessary.
                    */
            }
        }
    }    
}

package Drone_FeatureModel {
    
    import MBPLE::*;
    
    #feature drone {    
        #mandatoryFeature numberOfEngines {
            #altFeature fourEngines;
            #altFeature sixEngines;
        }
        
        #mandatoryFeature batteryKind {
            #altFeature standardBattery;
            #altFeature longDistanceBattery;
        }
    } 

    view DroneFeatureTree {
        expose drone::**;
        render Views::asTreeDiagram;
    }
}
```

# SysML extension for MBPLE

The MBPLE extension provides concepts for modeling features with SysML v2.

```sysml
library package MBPLE {
    
    requirement def Feature {
        attribute selected : ScalarValues::Boolean = false;
    }
    requirement features[*] : Feature;
    requirement alternativeFeatures[*] :> features {
        assert constraint {
            CollectionFunctions::size(alternativeFeatures->ControlFunctions::select { in f : Feature; f.selected == true }) == 1
        }
    }    
    requirement mandatoryFeatures[*] :> features {
        :>> selected = true;
        assert constraint { selected == true }
    }
    
    
    metadata def <feature> FeatureDSM :> Metaobjects::SemanticMetadata {
        :>> baseType = features meta SysML::Usage;
    }
    metadata def <altFeature> AlternativeFeatureDSM :> Metaobjects::SemanticMetadata {
        :>> baseType = alternativeFeatures meta SysML::Usage;
    }
    metadata def <mandatoryFeature> MandatoryFeatureDSM :> Metaobjects::SemanticMetadata {
        :>> baseType = mandatoryFeatures meta SysML::Usage;
    }    
}
```
